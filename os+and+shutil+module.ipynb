```python
# in depth analysis of file and folder
fileitr=os.walk(r'D:\\')
for current_path,folder_name,file_names in fileitr:
    print(f'current path :{current_path}')
    print(f'folder name:{folder_name}')
    print(f'file name :{file_names}')
```
